```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df =pd.read_csv('creditcard.csv')
df.head()
#all v1..n are the features of credit card..using pca method data converted to decimal
```

```python
df.info()
```

```python
#checking missing values in data points
df.isnull().sum()
```

```python
#distribution of legit and fraud transcation
df['Class'].value_counts()
```

This data is highly imbalanced
0---> normal transcations
1---> fraud  ""

```python
#separating data for analysis
legit=df[df.Class==0]
fraud=df[df.Class==1]
 
```

```python
print('legit',legit.shape)
print('fraud',fraud.shape)
```

```python
#stats
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#Compare the values for both transcations
df.groupby('Class').mean()
```

to manage highly unbalanced data we use
UNDER SAMPLING
Build a sample dataset cointaining similar distribution of legit and fraud transcation
number of fraud transcations ====>492


```python
legit_sample=legit.sample(n=492)
```

concatinating 2 dfs (so can make balanced data set)

```python
new_df=pd.concat([legit_sample,fraud],axis=0)
#axis=0 means concate row after row
#axis =1 means concate column by column
```

```python
new_df.tail()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

splitting the data into features and targets

```python
x=new_df.drop(columns='Class',axis=1)
y=new_df['Class']
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
#stratify means splitting data  evenly
```

model training

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
#accuracy score
x_train_prediction=model.predict(x_train)
data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print('training data accuracy :',data_accuracy)
```

```python
x_test_prediction=model.predict(x_test)
test_data_prediction= accuracy_score(x_test_prediction,y_test)
```

```python
print('test data accuracy score :',test_data_prediction)
```
